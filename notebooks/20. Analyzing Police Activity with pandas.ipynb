# Exploratory Data Analysis in Python

**Descripción del curso**

Ahora que ha aprendido los fundamentos de los pandas, este curso le dará la oportunidad de aplicar ese conocimiento respondiendo preguntas interesantes sobre un conjunto de datos real. Explorará el conjunto de datos del Stanford Open Policing Project y analizará el impacto del género en el comportamiento de la policía. Durante el curso, obtendrá más práctica en la limpieza de datos desordenados, la creación de visualizaciones, la combinación y remodelación de conjuntos de datos y la manipulación de datos de series de tiempo. Analizar la actividad policial con pandas le brindará una valiosa experiencia analizando un conjunto de datos de principio a fin, ¡lo que lo preparará para su carrera en ciencia de datos!
