## Aggregating Data

---

### Summary statistics

```python
homelessness.head()
```

```python
# ESTADISTICAS

print(f"Media: {homelessness['family_members'].mean():.2f}")
print(f"Max: {homelessness['family_members'].max()}")

# Otras estadisticas como
# .median() || .mode() || .min() || .max() ||
#      .var() || .std() || .sum() || .quantile()

def pct40(column):
    return column.quantile(0.4)
      
def pct30(column):
    return column.quantile(0.3)
      
print("----")
print(f"3er cuartil: {homelessness['family_members'].agg(pct30)}")
print("----")
print(homelessness["family_members"].agg([pct30,pct40]) )
      
# El método .agg () 
# permite aplicar sus propias funciones personalizadas a un DataFrame
# df['column'].agg(function)
```

```python
# print(homelessness["family_members"].cumsum())   # suma acumulada
# print(homelessness["family_members"].cummax())   # max comparacion
# print(homelessness["family_members"].cummin())   # min comparacion
# print(homelessness["family_members"].cumprod())  # producto acumulado
```

```python
# Trabajar con datos sobre las tiendas walmart - 
# cadena de tiendas departamentales en los EEUU
```

```python
homelessness.head()
```

```python
# Import NumPy and create custom IQR function
import numpy as np

def iqr(column): # A custom IQR function
    return column.quantile(0.75) - column.quantile(0.25)
    
# Print IQR of the temperature_c column
print(homelessness['family_members'].agg(iqr))
print("----")

# Update to print IQR of family_members, state_pop, & indiv_per_10k
print(homelessness[
    ["family_members", "state_pop", "indiv_per_10k"]].agg(iqr))
```

```python
# Update to print IQR and median of temperature_c, 
# fuel_price_usd_per_l, & unemployment
print(homelessness[
    ["family_members", "state_pop", "indiv_per_10k"]
    ].agg([iqr,np.median]))

# El método .agg () 
# facilita el cálculo de múltiples estadísticas en múltiples columnas,
# todo en una sola línea de código.
```

### Pivot tables

Las tablas dinámicas son otra forma de calcular estadisticas

```python
# homelessness.groupby("region")["state_pop"].mean()

# Otra forma  de groupBy
# pivot_table(
#      values:Nombre columna, 
#      index:Porque variable se va agrupar)

homelessness.pivot_table(values="state_pop", index="region") 
# Por default toma el valor medio
```

```python
# ESTADISTICA de resumen 
#     (aggfunc: pasarle una funcion en especifico)
homelessness.pivot_table(
    values="state_pop", index="region",aggfunc=np.median)
```

```python
# MULTIPLES estadisticas
homelessness.pivot_table(
    values="state_pop", index="region",aggfunc=[np.mean,np.median])
```

```python
homelessness.head()
```

```python
# homelessness.groupby(["region","state"])[["state_pop"]].mean() 
# -- state_pop medio y family_members medio 
#   agrupado por 2 variables "region","state"

# Otra forma
# Si margins=True la ultima fila y la columna 
# contienen la media de todos los valores (fila,columna)

# Hay valores NaN
homelessness.pivot_table(values="state_pop",
                         index="region",
                         columns="state", 
                         fill_value=0, 
                         margins=True)
```

### Creating DataFrames

```python
# Hay 2 formas de crear un diccionario:
#    (1)- De una lista de dictionarios => 
#         se contruye fila a fila
#    (2)- De un dictionarios de listas => 
#         se contruye columna a columna
```

```python
# (1)- De una lista de dictionarios
list_of_dicts = [ 
    {"name": "Ginger", "breed": "Dachshund", "height_cm": 22,
     "weight_kg": 10, "date_of_birth": "2019-03-14"},    
    {"name": "Scout", "breed": "Dalmatian", "height_cm": 59, 
     "weight_kg": 25, "date_of_birth": "2019-05-09"}
]
new_dogs = pd.DataFrame(list_of_dicts)
display(new_dogs)
```

```python
# (2)- De un dictionarios de lista
dict_of_lists = {
    "name": ["Ginger", "Scout"],
    "breed": ["Dachshund", "Dalmatian"],
    "height_cm": [22, 59],
    "weight_kg": [10, 25],
    "date_of_birth": ["2019-03-14", "2019-05-09"]
}
new_dogs2 = pd.DataFrame(dict_of_lists)
display(new_dogs2)
```

```python
airline_bumping = pd.read_csv("../datasets/airline_bumping.csv")
print(airline_bumping.head())

airline_totals = airline_bumping.groupby("airline")[
    ["nb_bumped", "total_passengers"]].sum()

airline_totals["bumps_per_10k"] = airline_totals["nb_bumped"] / airline_totals["total_passengers"] * 10000
print(airline_totals)
```

```python
airline_totals_sorted = airline_totals.sort_values(
    ["bumps_per_10k"], ascending=False)

print(airline_totals_sorted)

airline_totals_sorted.to_csv("../datasets/airline_totals_sorted.csv")
```
