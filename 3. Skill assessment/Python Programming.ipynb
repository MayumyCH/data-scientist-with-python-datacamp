```python
def count_words(filename):
  """Count the number of words in a file"""
  try:
    with open(filename, encoding='utf-8') as f:
      contents = f.read()
  except FileNotFoundError:
    print("The file {} cannot be found".format(filename))
  else:
    words = contents.split()
    num_words = len(words)
    print("The file {} has about {} words.".format(filename, num_words))
    
filename = "peter_rabbit.txt"
count_words(filename)
```

```python
def factorial(n):
    """returns n!"""
    return 1 if n < 2 else n * factorial(n-1)

factorial.__doc__
```

```python
# Cree una clase llamada Planet que tenga un atributo name y diameter_km que debe establecerse cuando se crea un nuevo objeto.
class Planet:
    def __init__(self, name, diameter_km):
        self.name = name
        self.diameter_km = diameter_km

    def woof(self):
        return 'woof!'
        
e = Planet("Earth", 12742)
print(e.name, e.diameter_km)
print(e.woof())
```
